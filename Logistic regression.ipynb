#  Import libraries


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
import pickle
```

#  Data preprocessing


```python
df=df.drop('id',axis=1)
df
```

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
df.isnull().sum()
```

# Remove columns with a correlation coefficient less than 0.2

```python
df['diagnosis'] = (df['diagnosis'] == 'M').astype(int)

correlation_matrix = df.corr()


selected_columns = correlation_matrix.columns[correlation_matrix.iloc[0] > 0.2]


df_filtered = df[selected_columns]
df_filtered
```

# Define dependent variable and independent variable

```python

x = df_filtered.drop('diagnosis',axis=1)
y = df['diagnosis']
print(x.shape)
print(y.shape)
```

```python

X, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
#x_test = np.concatenate((np.ones((len(y_test), 1)), x_test), axis=1)
#X = np.concatenate((np.ones((len(y_train), 1)), x_train), axis=1)
print(X.shape)
print(x_test.shape)
```

# Normalization


```python
scaler = StandardScaler() #create an instance of standard scaler
scaler.fit(X) # fit it to the training data

X = scaler.transform(X) #transform training data
x_test = scaler.transform(x_test) #transform validation data
```

# Logistic Regression from scratch


```python
class Logistic:
    #sigmoid function
    def _sigmoid(self, x):
        return 1 / (1 + np.exp(-x))
    def __init__(self, learning_rate=0.01, num_iterations=100):
        self.learning_rate = learning_rate
        self.num_iterations = num_iterations
        self.weights = None
        self.bias = None
    
    def fit(self, X, y):
        num_samples, num_features = X.shape
        self.weights = np.zeros(num_features)
        self.bias = 0
        
        # Gradient Descent
        for _ in range(self.num_iterations):
            linear_model = np.dot(X, self.weights) + self.bias
            y_predicted = self._sigmoid(linear_model)

            dw = (1 / num_samples) * np.dot(X.T, (y_predicted - y))
            db = (1 / num_samples) * np.sum(y_predicted - y)

            self.weights -= self.learning_rate * dw
            self.bias -= self.learning_rate * db
    
    def predict(self, X):
        linear_model = np.dot(X, self.weights) + self.bias
        y_predicted = self._sigmoid(linear_model)
        y_predicted_cls = [1 if i > 0.5 else 0 for i in y_predicted]
        return np.array(y_predicted_cls)

    
model = Logistic(learning_rate=0.01, num_iterations=2200)
model.fit(X, y_train)

# Make predictions"
Predictions1= model.predict(x_test)
accuracy = accuracy_score(y_test, Predictions1)
print(f'the model accuracy: {accuracy}')
```

# Using scikit-learn
